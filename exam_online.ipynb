```python
qa = pd.read_csv('500__questions.csv',encoding="cp1252")
```

```python
qa['question'] = qa['question'].apply(lambda x: " ".join(x.lower() for x in x.split()))
qa['question'].head()
```

```python
qa['question'] = qa['question'].str.replace('[^\w\s]','')
qa['question'].head()
```

```python
q=qa['question'].tolist()
```

```python
import sklearn
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.metrics.pairwise import euclidean_distances
```

```python
vectorizer = CountVectorizer()
```

```python
feature = vectorizer.fit_transform(q).todense()
```

```python
print(vectorizer.vocabulary_)
```

```python
aa = []
def call(an):
##    aa.append(a)
    p=sum(an)

i=0
while i<500:
         sum = 0 
         for f in feature :
           
            an =  euclidean_distances(feature[i],f);
            sum = an+sum
            avg = sum/500
            avg = avg.tolist()
            
##            print(sum)
##            call(an)
         aa.append(avg)
         i=i+1
      
      
```

```python
frame = pd.DataFrame(aa)
```

```python
fin = pd.concat([qa,frame], axis=1)
fin.head()
```

```python
fin.columns = ['Unnamed','question','average']
```

```python
ans_final = fin.sort_values(by = 'average')
ans_final
from pandas import ExcelWriter

writer = ExcelWriter('final_online.xlsx')
ans_final.to_excel(writer,'Sheet5')
writer.save()
```
